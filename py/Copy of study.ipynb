```python
def safe_first(L):
    if not L:
        return None
    return L[0]


def safe_last(L):
    if not L:
        return None
    return L[-1]


def safe_middle(L):
    if not L:
        return None
    if len(L) <= 2:
        return []
    return L[1:-1]


L = eval(input().strip())
result = safe_middle(L)


import sys
sys.stdout.write(str(result))
```

```python
def has_duplicates(L):
    seen = set()
    for item in L:
        if item in seen:
            return True
        seen.add(item)
    return False


L = eval(input().strip())
result = has_duplicates(L)

import sys
sys.stdout.write(str(result))

```

```python
def is_perfect(n):
    if n <= 1:
        return False

    total = 1  
    i = 2
    while i * i <= n:
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
        i += 1

    return total == n
```

```python
def is_perfect(n):
    if n <= 1:
        return False

    total = 1
    i = 2
    while i * i <= n:
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
        i += 1

    return total == n


n = int(input().strip())
result = is_perfect(n)

import sys
sys.stdout.write(str(result))

```
