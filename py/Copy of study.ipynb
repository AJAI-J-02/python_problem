```python
marks=[70,80,90]
backup=marks
backup[0]=100
print(marks)
print(backup)
```

```python
marks=[[70,80],[90,95]]
backup=marks.copy()

backup[0][0]=100
print(marks)
print(backup)
```

```python
marks=[[1,2],[3,4]]
backup=marks[:]
backup[1][1]=90
print(marks)
print(backup)
```
